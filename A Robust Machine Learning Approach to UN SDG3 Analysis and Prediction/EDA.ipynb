```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("SGD3_Standardized_data.csv")
df.head()
```

```python
df.dtypes
```

# Objective1: Analyze a comprehensive set of SDG3 indicators, unveiling patterns and trends.

```python
# Data summary
# Specify the columns to be imputed 
columns_to_impute = df.columns[4:12]

# Calculate the summary statistics
summary = df[columns_to_impute].describe()

# Format the summary statistics
summary = summary.applymap(lambda x: f'{x:.6f}')

print("\nDescriptive statistics:")
summary
```

```python
# Selecting the columns containing indicators and pc"s for relationship analysis
indicator_columns = df.columns[4:12]

# Calculate the correlation matrix
correlation_matrix = df[indicator_columns].corr()
#correlation_matrix
# Create a heatmap to visualize the correlations
plt.figure(figsize=(17, 14))
sns.heatmap(correlation_matrix, annot=True, cmap='mako', linewidths=1, square=True, fmt='.2f', annot_kws={'size': 20, 'weight': 'bold'}, vmin=-1, vmax=1)
plt.title('Correlation Heatmap of SDG3 Indicators', fontsize=30, fontweight='bold')
plt.xticks(fontweight='bold', fontsize=18)
plt.yticks(fontweight='bold', fontsize=18)
plt.show()
```

```python
# Histograms
df[indicator_columns].hist(figsize=(12, 12))
plt.tight_layout()
plt.show()
```

# SDG3 Progress and Performance Analysis:

```python

#Analysis of the SDG3 composite index over time to identify any trends or 
#changes in SDG3 progress and performance.

# Group the data by year and calculate the mean composite index for each year
mean_composite_index = df.groupby('year')['SDG3 Composite Index'].mean()

# Plot the mean composite index over time
plt.plot(mean_composite_index.index, mean_composite_index.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Mean SDG3 Composite Index')
plt.title('SDG3 Progress and Performance Over Time')
plt.show()
```

```python


# Group the data by year and calculate the mean composite index for each year
mean_composite_index = df.groupby('year')['SDG3 Composite Index'].mean()

# Create a dataframe for the mean composite index
mean_composite_index_df = pd.DataFrame({'Year': mean_composite_index.index, 'Mean Composite Index': mean_composite_index.values})

# Print the resulting dataframe
mean_composite_index_df
```

# Impact Analysis

```python
import pandas as pd


# Select the indicator columns
indicator_columns = df.columns[4:-1]  # Adjust the column range based on your dataset

# Calculate the correlation between each indicator and the SDG3 composite index
correlations = df[indicator_columns].corrwith(df['SDG3 Composite Index'])

# Sort the correlations in descending order
sorted_correlations = correlations.sort_values(ascending=False)

# Print the indicators with the highest positive impact
top_positive_indicators = sorted_correlations.head(6)  # Adjust the number of indicators to display
print("Top Indicators with Positive Impact:")
print(top_positive_indicators)

# Print the indicators with the highest negative impact
top_negative_indicators = sorted_correlations.tail(2)  # Adjust the number of indicators to display
print("\nTop Indicators with Negative Impact:")
print(top_negative_indicators)
```

```python
import matplotlib.pyplot as plt

# Select the indicator columns
indicator_columns = df.columns[4:-1]  # Adjust the column range based on your dataset

# Calculate the correlation between each indicator and the SDG3 composite index
correlations = df[indicator_columns].corrwith(df['SDG3 Composite Index'])

# Sort the correlations in descending order
sorted_correlations = correlations.sort_values(ascending=False)

# Plot the indicators with the highest positive impact
top_positive_indicators = sorted_correlations.head(6)  # Adjust the number of indicators to display
plt.bar(top_positive_indicators.index, top_positive_indicators)
plt.xlabel('Indicator', fontsize=12, fontweight='bold')
plt.ylabel('Correlation with SDG3 Composite Index', fontsize=12, fontweight='bold')
plt.title('Top Indicators with Positive Impact on SDG3', fontsize=15, fontweight='bold')
plt.xticks(rotation=45)
plt.show()

# Plot the indicators with the highest negative impact
top_negative_indicators = sorted_correlations.tail(2)  # Adjust the number of indicators to display
plt.bar(top_negative_indicators.index, top_negative_indicators)
plt.xlabel('Indicator', fontsize=12, fontweight='bold')
plt.ylabel('Correlation with SDG3 Composite Index', fontsize=12, fontweight='bold')
plt.title('Top Indicators with Negative Impact on SDG3', fontsize=15, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```
