## Handling the Missing values

```python

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Create a copy of the dataframe for imputation
df_imputed = df.copy()

# Specify the columns to be imputed 
columns_to_impute = df_imputed.columns[4:]

# Initialize the IterativeImputer
iterimp = IterativeImputer()

# Fit the imputer to the selected columns and impute missing values
df_imputed[columns_to_impute] = iterimp.fit_transform(df_imputed[columns_to_impute])

```

# creating/quantifying SDG3 progress representation

```python
from sklearn.preprocessing import StandardScaler


# Selecting the indicator columns
indicator_columns = df_imputed.columns[4:]

# Normalizing the indicator columns
scaler = StandardScaler()
df_imputed[indicator_columns] = scaler.fit_transform(df_imputed[indicator_columns])

df_imputed.head()
```

## Calculate Distance Measures.

```python
from sklearn.metrics.pairwise import euclidean_distances

# Select the indicator columns
indicator_columns = df_imputed.columns[4:]

# Calculate the Euclidean distance matrix
distance_matrix = euclidean_distances(df_imputed[indicator_columns])

# Print the distance matrix
distance_matrix
```

# Composite Index Calculation:

```python
# Calculate the average of the distance measures
composite_index = distance_matrix.mean(axis=1)

# Create a new column for the composite index in the dataframe
df_imputed['SDG3 Composite Index'] = composite_index

# Print the resulting dataframe with the composite index
df_imputed.head()
```

```python
# save the Standardized dataset

df_imputed.to_csv('SGD3_Standardized_data.csv', index=False)
```
